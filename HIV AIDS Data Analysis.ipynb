## Data Analysis and Visualization

### Number of HIV cases among adults (19-45)


#### Region-wise:

```python
regional_cases = adult_df.groupby(['WHO Region','Year']).mean()['Count_median']
```

```python
regional_cases
```

```python
regional_cases.plot(kind="bar", subplots=True, figsize=(30,30))
```

```python
regional_cases.unstack(level=1).plot(kind="bar", subplots=True, figsize=(30,30))
```

```python
regional_cases.unstack(level=0).plot(kind="bar", subplots=True, figsize=(30,30))
```

##### \* From the above different plots, it is clear that the Africa has shown a steady decrease in the number of HIV cases in the past decade (2000 to 2018). But, European countries show gradual increase in the number of HIV cases since the decade. Other regions happen to have the relatively same amount of cases for the past decade.\*

#### \*However, Africa tends to have higher number of cases till now.\*

### Country-wise:

```python
countrywisecases = adult_df.groupby(['Country','Year']).mean()['Count_median']
```

```python
countrywisecases.unstack(level=1).plot(kind="bar", subplots=True, figsize=(30,20))
```

##### \*Eswatini, in South Africa, is the place which holds the tragically highest number of HIV AIDS cases now. However, Botswana which had the highest number of cases in 2000, the start of the decade, gradually reduced through the decade but still stands in 3rd position. Besides, Lesotho holds the second highest number of cases throughout the decade.\*

### Number of deaths due to HIV AIDS

#### Region-wise:

```python
regional_deaths = death_df.groupby(['WHO Region', 'Year']).mean()['Count_median']
```

```python
regional_deaths.unstack(level=0).plot(kind="bar", subplots=True, figsize=(20,20))
```

```python
regional_deaths.unstack(level=1).plot(kind="bar", subplots=True, figsize=(20,20))
```

#### \*Africa has the highest number of deaths due to HIV AIDS throughout the decade. The number of deaths in South-East Asia has gradually increased from start of the decade and stands on the second highest position of having larger number of deaths. The growth in number of deaths in Western Pacific and Europe throughout the decade is also significant.\*

### Country-wise:

```python
countrywisedeaths = death_df.groupby(['Country','Year']).mean()['Count_median']
```

```python
countrywisedeaths.unstack(level=1).plot(kind="bar", subplots=True, figsize=(30,20))
```

##### \*In the start of the decade during 2000, Kenya and Zimbabwe holds the greatest amount of deaths due to HIV AIDS. But, their amount reduced greatly throughout the decade. However, South Africa again holds the highest number of deaths due to HIV AIDS throughout the decade. Other significant increase in number of deaths are shown in Nigeria, Indonesia and Mozambique.\*


## Number of people living with HIV AIDS

### Region-wise

```python
peoplewithhiv = pd.DataFrame(hiv_living_df.groupby(['WHO Region', 'Year']).mean()['Count_median'])
peoplewithhiv
```

```python
plt.figure(figsize=(10,6))
plt.title("People living with HIV regional")
sns.barplot(x=peoplewithhiv.index.get_level_values(0), y='Count_median', data=peoplewithhiv, hue=peoplewithhiv.index.get_level_values(1))
plt.xticks(rotation=90)
plt.ylabel("People with HIV count")
```

##### \* Even here, Africa seems to have the highest number of people living with HIV AIDS since the start of the decade and still holds the first position with nearly 6 lakh people living with HIV AIDS. Secondly, South-East Asia tend to have a steady increase in number of people who live with HIV AIDS. Americas, eventhough they showed slight increase in the beginning of the decade, has managed to reduce it significantly at the end of the decade(2018). Other regions show gradual increase in number of people living with HIV.\*

### Country-wise:

```python
countrywise_HIVpeople = hiv_living_df.groupby(['Country','Year']).mean()['Count_median']
```

```python
countrywise_HIVpeople.unstack(level=1).plot(kind="bar", subplots=True, figsize=(25,25))
```

#### \* From the above plot, it is clear that South Africa has the highest number of people living with HIV AIDS throughout the decade.\*

## Prevention of mother-to-child transmission estimate

```python
mother2child_prevention['Received Antiretrovirals'].value_counts()
```

#### \*There is the value "No Data" which poses no real meaning. So, the value is replaced with 0.\*

```python
mother2child_prevention['Received Antiretrovirals'].replace("No data", 0, inplace=True)
mother2child_prevention['Received Antiretrovirals'] = pd.to_numeric(mother2child_prevention['Received Antiretrovirals'], errors="coerce")
mother2child_prevention['Received Antiretrovirals'] = mother2child_prevention['Received Antiretrovirals'].astype(float)
```

```python
preventionofTransmission = pd.DataFrame(mother2child_prevention.groupby("WHO Region").mean()['Received Antiretrovirals'])
preventionofTransmission
```

```python
plt.figure(figsize=(10,6))
plt.title("Pregnant women with HIV who received antiretrovirals - Regional")
sns.barplot(x=preventionofTransmission.index, y='Received Antiretrovirals', data=preventionofTransmission, hue=preventionofTransmission.index)
plt.xticks(rotation=90)
plt.ylabel("Number of pregnant women with HIV who recieved antiretrovirals")
```

### From the above plot, it is clear that Africa has the highest number of pregnant women who received antiretrovirals. This is meaningful since the Africa holds the highest position of all the analysis done since(in this project).

```python
needAR_women = pd.DataFrame(mother2child_prevention.groupby('WHO Region').mean()['Needing antiretrovirals_median'])
needAR_women
```

```python
plt.figure(figsize=(10,6))
plt.title("Pregnant women with HIV who needs antiretrovirals - Regional")
sns.barplot(x=needAR_women.index, y='Needing antiretrovirals_median', data=needAR_women, hue=needAR_women.index)
plt.xticks(rotation=90)
plt.ylabel("Number of pregnant women with HIV who needs antiretrovirals")
```

### From the above plot, it is clear that Africa has the highest number of women who needs antiretrovirals to prevent transmission. To be noted, Africa has the number of women who needs antiretrovirals **greater** than number of women who received antiretrovirals.

## ART (Anti Retro-viral Therapy) coverage among people living with HIV estimates:

```python
artCoverage_df['Reported number of people receiving ART'].value_counts()
```

##### \* Even here, the value "Nodata" is of no meaning, so it is replaced with 0. \*

```python
artCoverage_df['Reported number of people receiving ART'].replace("Nodata", 0, inplace=True)
artCoverage_df['Reported number of people receiving ART'] = pd.to_numeric(artCoverage_df['Reported number of people receiving ART'], errors="coerce")
artCoverage_df['Reported number of people receiving ART'] = artCoverage_df['Reported number of people receiving ART'].astype(float)
```

```python
PeoplereceivingART = pd.DataFrame(artCoverage_df.groupby('WHO Region').mean()['Reported number of people receiving ART'])
PeoplereceivingART
```

```python
plt.figure(figsize=(10,6))
plt.title("People who receives ART - Regional")
sns.barplot(x=PeoplereceivingART.index, y='Reported number of people receiving ART', data=PeoplereceivingART, hue=PeoplereceivingART.index)
plt.xticks(rotation=90)
plt.ylabel("Number of people who receives ART")
```

#### Africa tends to have the greatest number of people who are receiving ART. Other regions have significantly low.

```python
ARTcoverage = pd.DataFrame(artCoverage_df.groupby('WHO Region').mean()['Estimated ART coverage among people living with HIV (%)_median'])
ARTcoverage
```

```python
plt.figure(figsize=(10,10))
plt.title("ART coverage of people living with HIV - Regional")
sns.barplot(x=ARTcoverage.index, y='Estimated ART coverage among people living with HIV (%)_median', data=ARTcoverage, hue=ARTcoverage.index)
plt.xticks(rotation=90)
plt.ylabel("Estimated ART coverage")
plt.legend(loc="upper center", bbox_to_anchor=(1, 1.2))
```

#### Although the number of people receiving ART is more in Africas, average % wise is high in Europe and Western Pacific.

## ART (Anti Retro-viral Therapy) coverage among children estimates:

```python
pediatricCoverage_df['Reported number of children receiving ART'].value_counts()
```

##### \* The value "No data" is replaced with 0. \*

```python
pediatricCoverage_df['Reported number of children receiving ART'].replace("No data", 0, inplace=True)
pediatricCoverage_df['Reported number of children receiving ART'] = pd.to_numeric(pediatricCoverage_df['Reported number of children receiving ART'], errors="coerce")
pediatricCoverage_df['Reported number of children receiving ART'] = pediatricCoverage_df['Reported number of children receiving ART'].astype(float)
```

```python
ChildrenreceivingART = pd.DataFrame(pediatricCoverage_df.groupby('WHO Region').mean()['Reported number of children receiving ART'])
ChildrenreceivingART
```

```python
plt.figure(figsize=(10,6))
plt.title("Chidren who receives ART - Regional")
sns.barplot(x=ChildrenreceivingART.index, y='Reported number of children receiving ART', data=ChildrenreceivingART, hue=ChildrenreceivingART.index)
plt.xticks(rotation=90)
plt.ylabel("Number of children who receives ART")
```

#### Africa has the highest number of children who receives ART, followed by South-East Asia and Western Pacific.

```python
childrenneedingART = pd.DataFrame(pediatricCoverage_df.groupby('WHO Region').mean()['Estimated number of children needing ART based on WHO methods_median'])
childrenneedingART
```

```python
plt.figure(figsize=(10,6))
plt.title("Chidren who needs ART - Regional")
sns.barplot(x=childrenneedingART.index, y='Estimated number of children needing ART based on WHO methods_median', data=childrenneedingART, hue=childrenneedingART.index)
plt.xticks(rotation=90)
plt.ylabel("Number of children who needs ART")
```

#### Africa has the tremendously highest number of children who needs ART rather than those who are receiving it!!

```python
ChildARTcoverage = pd.DataFrame(pediatricCoverage_df.groupby('WHO Region').mean()['Estimated ART coverage among children (%)_median'])
ChildARTcoverage
```

```python
plt.figure(figsize=(10,6))
plt.title("Estimated ART coverage in children - Regional")
sns.barplot(x=ChildARTcoverage.index, y='Estimated ART coverage among children (%)_median', data=ChildARTcoverage, hue=ChildARTcoverage.index)
plt.xticks(rotation=90)
plt.ylabel("Estimated ART coverage")
plt.legend(loc="upper center", bbox_to_anchor=(1, 1.4), fancybox=True, shadow=True)
```

### Though the Africa has the highest number of children who needs ART, percent wise South-East Asia, Western Pacific and Europe has the highest values.
